```python
data['education']= data['education'].fillna(data['education'].median())
data['cigsPerDay']=data['cigsPerDay'].fillna(data['cigsPerDay'].mean())
data['BPMeds']=data['BPMeds'].fillna(data['BPMeds'].median())
data['totChol']=data['totChol'].fillna(data['totChol'].mean())
data['BMI']=data['BMI'].fillna(data['BMI'].mean())
data['heartRate']=data['heartRate'].fillna(data['heartRate'].mean())
data['glucose']=data['glucose'].fillna(data['glucose'].mean())
```

```python
data.isnull().sum()
```

```python
correlation_matrix = data.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
X = data[['male','age','education','BPMeds','cigsPerDay','prevalentStroke','prevalentHyp','diabetes','totChol','sysBP','diaBP','BMI','heartRate','currentSmoker','glucose']]
X.head()
y= data['TenYearCHD']
y.head()

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)
```

```python
print(type(X_train))
print(type(X_test))
print(type(y_train))
print(type(y_test))
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
y_predict = lr.predict(X_test)
type(y_predict)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Assuming y_test and y_predict are numpy arrays or lists containing continuous values
# Let's convert them to binary classes using a threshold
threshold = 0.5  # Adjust this threshold as needed

y_test_binary = np.where(y_test > threshold, 1, 0)
y_predict_binary = np.where(y_predict > threshold, 1, 0)

# Now, you can use these binary predictions for calculating classification metrics
accuracy = accuracy_score(y_test_binary, y_predict_binary)
precision = precision_score(y_test_binary, y_predict_binary)
recall = recall_score(y_test_binary, y_predict_binary)
f1 = f1_score(y_test_binary, y_predict_binary)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
